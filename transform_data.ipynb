```python
import pandas as pd
```

```python
ev_df = pd.read_csv('Resources/electric_reg_by_state.csv')
fuel_prices = pd.read_csv('Resources/fuel_prices_by_state.csv')
population = pd.read_csv('Resources/population.csv')
income = pd.read_csv('Resources/income.csv')
```

## Make a primary key

```python
def create_id(file):
    file['id'] = ""
    for x in range(50):
        file['id'][x] = x+1
    return file
```

```python
create_id(population)
create_id(fuel_prices)
create_id(ev_df)
create_id(income)
```

## Check if states match

```python
def matching(file_1, file_2, file_3, file_4):
    for x in range(50):
        if file_1['id'][x] == file_2['id'][x] == file_3['id'][x] == file_4['id'][x]:
            print('Matching')
        else:
            print(f"Error: {x} {file_1['id'][x]}")
            
```

```python
matching(fuel_prices, ev_df, population, income)
```

## Create state id table

```python
states = pd.DataFrame({'id':ev_df['id'],
                     'state':ev_df['state']})
```

## Set Index

```python
ev_df = ev_df.set_index('id')
fuel_prices = fuel_prices.set_index('id')
states = states.set_index('id')
population = population.set_index('id')
income = income.set_index('id')
```

## Export

```python
ev_df.to_csv('clean_data/ev_registrations.csv')
fuel_prices.to_csv('clean_data/fuel_prices_2020.csv')
states.to_csv('clean_data/states.csv')
population.to_csv('clean_data/population.csv')
income.to_csv('clean_data/income.csv')
```
